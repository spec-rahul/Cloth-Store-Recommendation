```python
df['Age'] = df['Age'].apply(lambda x:x.split())
#movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
df['Gender'] = df['Gender'].apply(lambda x:x.split())
#movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
df['Garment'] = df['Garment'].apply(lambda x:x.split())
#movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
df['Branded'] = df['Branded'].apply(lambda x:x.split())
```

```python
df['Price'] = df['Price'].apply(str)
df['Price'] = df['Price'].apply(lambda x:x.split())
```

```python
df['Rating'] = df['Rating'].apply(str)
df['Rating'] = df['Rating'].apply(lambda x:x.split())
```

```python
df['tags']=df['Age'] + df['Gender'] + df['Garment'] + df['Branded'] + df['Price'] + df['Rating']
```

```python
new_df = df.drop(columns=['Age','Gender','Garment','Branded','Price','Rating'])
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))

new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df.head()
```
